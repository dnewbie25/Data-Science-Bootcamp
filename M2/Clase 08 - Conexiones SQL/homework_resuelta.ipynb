Ahora se importa el csv en un dataframe

```python
import pandas as pd
datos = pd.read_csv('oferta_gastronomica.csv', delimiter=',', encoding='utf-8')
```

```python
tabla_para_cargar = datos[['id','nombre','categoria','direccion_completa','barrio','comuna']]
tabla_para_cargar.head()
```

```python
tabla_para_cargar.rename(columns={'id':'id_local','direccion_completa':'direccion'}, inplace=True)
tabla_para_cargar.head()
```

```python
tabla_para_cargar.info()
```

```python
tabla_para_cargar['id_local'].loc[1]
```

```python
# revisar video minuto 56:07, execute many y primero poner todo en tuplas
datos_tupla = []

for row in tabla_para_cargar.index:
  datos_tupla.append((tabla_para_cargar.loc[row]['id_local'],tabla_para_cargar.loc[row]['nombre'],tabla_para_cargar.loc[row]['categoria'],tabla_para_cargar.loc[row]['direccion'],tabla_para_cargar.loc[row]['barrio'],tabla_para_cargar.loc[row]['comuna']))

datos_tupla
```

```python
tabla_para_cargar=tabla_para_cargar.dropna()
tabla_para_cargar.info()
```

```python
cursor.executemany("""INSERT INTO oferta_gastronomica (id_local,nombre,categoria,direccion,barrio,comunca) VALUES (%s,%s,%s,%s,%s,%s)""",datos_tupla)
```

```python
cursor.execute('SELECT *  FROM oferta_gastronomica')
cursor.fetchall()
```

```python
# barrios con mayor cantidad de bares
cursor.execute('SELECT count(id_local) as pubs, barrio from henry.oferta_gastronomica where categoria="PUB" group by barrio order by pubs desc')
cursor.fetchone()
```

```python
# cantidad de locales por categoria
cursor.execute('select count(id_local) as pubs, categoria from henry.oferta_gastronomica group by categoria order by pubs;')
cursor.fetchall()
```

```python
cursor.execute("SELECT count(id_local) as restaurantes, comuna from henry.oferta_gastronomica where categoria='RESTAURANTE' group by comuna order by restaurantes desc")
cursor.fetchall()
```
